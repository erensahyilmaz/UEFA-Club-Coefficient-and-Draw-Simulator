```python
import matplotlib.pyplot as plt
import numpy as np

def plot_team_rankings(dfs, teams, colors=None, title="UEFA Club Coefficients"):
    """
    dfs: list of dataframeler, her biri 'Yıl' ve 'Sıra' sütunlarını içermeli
    teams: list of takım isimleri
    colors: list of renkler, teams ile aynı uzunlukta (default matplotlib renkleri)
    """
    if colors is None:
        # Eğer renk verilmezse matplotlib default renkleri
        colors = plt.cm.tab10.colors[:len(teams)]

    x_offset = 0
    y_offset = -0.5  # Y ekseni ters olduğu için yukarı kaydırmak negatif

    plt.figure(figsize=(12,6))

    # Arka plan renkleri
    ax = plt.gca()
    ax.set_facecolor("#403939")  # grafik alanı beyaz
    plt.gcf().patch.set_facecolor("lightgrey")  # en arka plan gri

    # Her takım için plot
    for df, team, color in zip(dfs, teams, colors):
        plt.plot(df["Yıl"], df["Sıra"], marker="o", color=color, label=team)
        for x, y in zip(df["Yıl"], df["Sıra"]):
            plt.text(x + x_offset, y + y_offset, str(y), fontsize=9,
                     ha='center', va='bottom', color="white")

    # Y ekseni ters (küçük sıra = üst sıra)
    plt.gca().invert_yaxis()

    # X eksenini 2 yılda bir gösterecek şekilde ayarla
    all_years = np.concatenate([df["Yıl"].values for df in dfs])
    years = np.arange(all_years.min(), all_years.max()+1, 2)
    plt.xticks(years, color="#362121")

    # Eksen adları ve başlık
    plt.xlabel("Year", color="#362121")
    plt.ylabel("Ranking Position", color="#362121")
    plt.title(title, color="#362121")

    # Izgara ve legend
    plt.grid(True, color="#362121", linestyle="--", alpha=0.5)
    plt.legend(facecolor="#736B6B", edgecolor="#362121", labelcolor="#362121",bbox_to_anchor=(0.5, 1.15), ncol=len(teams))

    # Y ekseni renkleri
    plt.gca().tick_params(axis="y", colors="#362121")

    plt.show()
```

```python
import pandas as pd

# Örnek: df_all daha önce birleştirildi
def rank_row(row):
    # Sadece float değerler üzerinde rank uygula, NaN'ları koru
    return row.rank(ascending=False, method="min")

df_ranks = df_all.copy()
df_ranks[["Beşiktaş","Fenerbahçe","Galatasaray"]] = \
    df_all[["Beşiktaş","Fenerbahçe","Galatasaray"]].apply(rank_row, axis=1)

# NaN değerleri 0 ile doldur ve int'e çevir
df_ranks[["Beşiktaş","Fenerbahçe","Galatasaray"]] = \
    df_ranks[["Beşiktaş","Fenerbahçe","Galatasaray"]].fillna(0).astype(int)

df_ranks
```

```python
import pandas as pd

# teams ve df_ranks hazır olmalı
teams = ["Beşiktaş", "Fenerbahçe", "Galatasaray"]
years = df_ranks["Yıl"].tolist()

# Boş tablo oluştur (sıralar 1-3)
table = pd.DataFrame(index=[1,2,3], columns=years)

# Her yıl için takımları sıraya yerleştir
for year in years:
    for team in teams:
        rank = df_ranks.loc[df_ranks["Yıl"] == year, team].values[0]
        if pd.isna(rank):
            continue
        rank = int(rank)
        table.at[rank, year] = team  # 1,2,3 sıralarına yerleştir

# Boş olan hücreleri '-' ile doldur
table = table.fillna('-')

table
```

```python


# --- 2️⃣ Her takımın kaç kere 1,2,3,4 ve yok olduğunu hesapla ---
summary = pd.DataFrame(index=teams, columns=[1,2,3])

for team in teams:
    # 1-4. sıralar
    for rank in [1,2,3]:
        summary.at[team, rank] = (df_ranks[team] == rank).sum()

summary
```

```python
def create_team_dfs_from_list(takimlar):
    for team, country in zip(df['Pot 1'].iloc[0:9].tolist(),ulkeler[0:9]):
        var_name = f"df_{team.replace(' ', '_').replace('-', '_')}"
        # 8 satırlık dataframe oluştur, Takım sütununu doldur, diğerleri boş
        globals()[var_name] = pd.DataFrame({
            'Takım': [team]*8,
            'Ülke' : [country]*8,
            'Pot' : ['Pot 1']*8,
            'Rakip': ['']*8,
            'Rakip Ülke' : ['']*8,
            'Rakip Pot' :['']*8,
            'Durum': ['']*8
        })
        print(f"Created dataframe: {var_name}")
    for team, country in zip(df['Pot 2'].iloc[0:9].tolist(), ulkeler[9:18]):
        var_name = f"df_{team.replace(' ', '_').replace('-', '_')}"
        globals()[var_name] = pd.DataFrame({
            'Takım': [team]*8,
            'Ülke': [country]*8,
            'Pot' : ['Pot 2']*8,
            'Rakip': ['']*8,
            'Rakip Ülke': ['']*8,
            'Rakip Pot' :['']*8,
            'Durum': ['']*8
        })
        print(f"Created dataframe: {var_name}")
 # Pot 3'deki takımlar
    for team, country in zip(df['Pot 3'].iloc[0:9].tolist(), ulkeler[18:27]):
        var_name = f"df_{team.replace(' ', '_').replace('-', '_')}"
        globals()[var_name] = pd.DataFrame({
            'Takım': [team]*8,
            'Ülke': [country]*8,
            'Pot' : ['Pot 3']*8,
            'Rakip': ['']*8,
            'Rakip Ülke': ['']*8,
            'Rakip Pot' :['']*8,
            'Durum': ['']*8
        })
        print(f"Created dataframe: {var_name}")
    for team, country in zip(df['Pot 4'].iloc[0:9].tolist(), ulkeler[27:36]):
        var_name = f"df_{team.replace(' ', '_').replace('-', '_')}"
        globals()[var_name] = pd.DataFrame({
            'Takım': [team]*8,
            'Ülke': [country]*8,
            'Pot' : ['Pot 4']*8,
            'Rakip': ['']*8,
            'Rakip Ülke': ['']*8,
            'Rakip Pot' :['']*8,
            'Durum': ['']*8
        })
        print(f"Created dataframe: {var_name}")

def create_pot_ave_from_df(df,ulkeler):
    # Pot 1
    pot1_teams = df['Pot 1'].iloc[0:9].tolist()
    pot1_rows = []
    i=0
    for team in pot1_teams:
        pot1_rows.append({'Takım': team, 'Ülke': ulkeler[i],'Durum': 'İç saha'})
        pot1_rows.append({'Takım': team, 'Ülke': ulkeler[i],'Durum': 'Deplasman'})
        i=i+1
    global pot1_ave
    pot1_ave = pd.DataFrame(pot1_rows)
    print("Created pot1_ave")

    # Pot 2
    pot2_teams = df['Pot 2'].iloc[0:9].tolist()
    pot2_rows = []
    for team in pot2_teams:
        pot2_rows.append({'Takım': team, 'Ülke': ulkeler[i],'Durum': 'İç saha'})
        pot2_rows.append({'Takım': team, 'Ülke': ulkeler[i],'Durum': 'Deplasman'})
        i=i+1
    global pot2_ave
    pot2_ave = pd.DataFrame(pot2_rows)
    print("Created pot2_ave")

    # Pot 3
    pot3_teams = df['Pot 3'].iloc[0:9].tolist()
    pot3_rows = []
    for team in pot3_teams:
        pot3_rows.append({'Takım': team, 'Ülke': ulkeler[i],'Durum': 'İç saha'})
        pot3_rows.append({'Takım': team, 'Ülke': ulkeler[i],'Durum': 'Deplasman'})
        i=i+1
    global pot3_ave
    pot3_ave = pd.DataFrame(pot3_rows)
    print("Created pot3_ave")

    # Pot 4
    pot4_teams = df['Pot 4'].iloc[0:9].tolist()
    pot4_rows = []
    for team in pot4_teams:
        pot4_rows.append({'Takım': team, 'Ülke': ulkeler[i],'Durum': 'İç saha'})
        pot4_rows.append({'Takım': team, 'Ülke': ulkeler[i],'Durum': 'Deplasman'})
        i=i+1
    global pot4_ave
    pot4_ave = pd.DataFrame(pot4_rows)
    print("Created pot4_ave")


def create_pot_fix():
    global pot_matches

    pot_matches = pd.DataFrame(columns=['Takım','Ülke','Pot','Rakip','Rakip Ülke','Rakip Pot','Durum'])

    print("Created matches")

# Örnek kullanım
def pot1_draw(team_name, team_df):
    global pot_matches,pot1_ave
    n = 2
    pot_val = team_df['Pot'].iloc[0]  # 'Pot 1' / 'Pot 2' / ...
    # Daha önce eklenmiş maçları bul
    matches = pot_matches[pot_matches['Rakip'] == team_name] if not pot_matches.empty else pd.DataFrame()
    if not matches.empty :
        matches = matches[matches['Pot'] == 'Pot 1']
        for idx, row in matches.iterrows():
            reverse_status = 'Deplasman' if row['Durum'] == 'İç saha' else 'İç saha'

            # team_country'i güvenli al
            team_row = team_df.loc[team_df['Takım'] == team_name, 'Ülke']
            team_country = team_row.iloc[0] if not team_row.empty else "Unknown"

            # rakip ve rakip_ulke (pot_matches'ten alınıyor)
            rakip = row['Takım']
            rakip_row = pot_matches.loc[pot_matches['Takım'] == rakip, 'Ülke']
            rakip_ulke = rakip_row.iloc[0] if not rakip_row.empty else "Unknown"


            # durum
            durum = reverse_status

            # satırı ekle
            empty_idx = team_df[team_df['Rakip'] == ''].index.min()
            if pd.notna(empty_idx):
                team_df.loc[empty_idx] = {'Takım': team_name,'Ülke' : team_country,'Pot':pot_val,
                                          'Rakip': rakip, 'Rakip Ülke': rakip_ulke, 'Rakip Pot': 'Pot 1',
                                          'Durum': durum}
            else:
                team_df.loc[len(team_df)] = {'Takım': team_name,'Ülke' : team_country,'Pot':pot_val,
                                            'Rakip': rakip, 'Rakip Ülke': rakip_ulke, 'Rakip Pot': 'Pot 1',
                                            'Durum': durum}
    n = 2 - len(matches) if not matches.empty else 2

    # 2️⃣ Yeni eşleşmeler seç
    while n > 0 and not pot1_ave.empty:
        # 1️⃣ Önce pot_matches içinde bu takım var mı kontrol et
        team_fix = pot_matches[(pot_matches['Takım'] == team_name) & (pot_matches['Rakip Pot'] == 'Pot 1')]
        opponent_fix = pot_matches[(pot_matches['Rakip'] == team_name) & (pot_matches['Rakip Pot'] == 'Pot 1')]
        opponent_pot=team_df[team_df['Rakip Pot'] == 'Pot 1']
        # team_df ve pot1_ave her satırda artık 'Ülke' sütununa sahip
        team_country = team_df['Ülke'].iloc[0]  # team_name'in ülkesi

        # Önce mevcut rakipleri ve rakip ülkelerini al
        existing_opponents = team_df['Rakip'].tolist()
        # Rakip olarak seçilmiş ülkelerin sayısını hesapla
        country_counts = team_df['Rakip Ülke'].value_counts()

        # 2'den fazla rakip olan ülkeleri blokla
        blocked_countries = [country for country, count in country_counts.items() if count >= 2]


        if not team_fix.empty:
            current_status = team_fix.iloc[0]['Durum']
            desired_status = 'Deplasman' if current_status == 'İç saha' else 'İç saha'

            valid_choices = pot1_ave[
                (pot1_ave['Takım'] != team_name) &
                (pot1_ave['Durum'] == desired_status) &
                (~pot1_ave['Takım'].isin(existing_opponents)) &
                (pot1_ave['Ülke'] != team_country) &
                (~pot1_ave['Ülke'].isin(blocked_countries))
            ]

        elif not opponent_fix.empty:
            current_status = opponent_fix.iloc[0]['Durum']
            desired_status = current_status

            valid_choices = pot1_ave[
                (pot1_ave['Takım'] != team_name) &
                (pot1_ave['Durum'] == desired_status) &
                (~pot1_ave['Takım'].isin(existing_opponents)) &
                (pot1_ave['Ülke'] != team_country) &
                (~pot1_ave['Ülke'].isin(blocked_countries))
            ]

        else:
            valid_choices = pot1_ave[
                (pot1_ave['Takım'] != team_name) &
                (~pot1_ave['Takım'].isin(existing_opponents)) &
                (~((pot1_ave['Durum'] == 'İç saha') & (opponent_pot['Durum'].eq('İç saha')).any())) &
                (~((pot1_ave['Durum'] == 'Deplasman') & (opponent_pot['Durum'].eq('Deplasman')).any())) &
                (pot1_ave['Ülke'] != team_country) &
                (~pot1_ave['Ülke'].isin(blocked_countries))
            ]



        if valid_choices.empty:
          print(f"⚠️ {team_name} için uygun rakip kalmadı.")
          raise RestartDraw  # Burada özel exception fırlatıyoruz


        idx = valid_choices.sample(n=1).index[0]
        selected = pot1_ave.loc[idx]

        rakip = selected['Takım']
        durum = selected['Durum']
        rakip_ulke = selected['Ülke']


        # İlk boş satıra yaz
        empty_idx = team_df[team_df['Rakip'] == ''].index.min()
        if pd.notna(empty_idx):
            team_df.loc[empty_idx] = {'Takım': team_name,'Ülke' : team_country,'Pot':pot_val, 'Rakip': rakip, 'Rakip Ülke': rakip_ulke ,'Rakip Pot': 'Pot 1','Durum': durum}
        else:
            team_df.loc[len(team_df)] = {'Takım': team_name, 'Ülke' : team_country,'Pot':pot_val, 'Rakip': rakip, 'Rakip Ülke': rakip_ulke ,'Rakip Pot': 'Pot 1','Durum': durum}

        # pot_matches’e ekle
        pot_matches.loc[len(pot_matches)] = {'Takım': team_name, 'Ülke' : team_country,'Pot':pot_val, 'Rakip': rakip, 'Rakip Ülke': rakip_ulke ,'Rakip Pot': 'Pot 1','Durum': durum}

        # Seçilen satırı sil
        pot1_ave.drop(idx, inplace=True)
        # 3️⃣ Takımın kendi adını ve çekilen durumun tersini sil
        reverse_status = 'Deplasman' if durum == 'İç saha' else 'İç saha'
        pot1_ave = pot1_ave.drop(
            pot1_ave[(pot1_ave['Takım'] == team_name) & (pot1_ave['Durum'] == reverse_status)].index
        )

        n -= 1

def pot2_draw(team_name, team_df):
    global pot_matches, pot2_ave
    n = 2
    pot_val = team_df['Pot'].iloc[0]
    # Daha önce eklenmiş maçları bul
    matches = pot_matches[pot_matches['Rakip'] == team_name] if not pot_matches.empty else pd.DataFrame()
    if not matches.empty :
        matches = matches[matches['Pot'] == 'Pot 2']
        for idx, row in matches.iterrows():
            reverse_status = 'Deplasman' if row['Durum'] == 'İç saha' else 'İç saha'

            # team_country'i güvenli al
            team_row = team_df.loc[team_df['Takım'] == team_name, 'Ülke']
            team_country = team_row.iloc[0] if not team_row.empty else "Unknown"

            # rakip ve rakip_ulke (pot_matches'ten alınıyor)
            rakip = row['Takım']
            rakip_row = pot_matches.loc[pot_matches['Takım'] == rakip, 'Ülke']
            rakip_ulke = rakip_row.iloc[0] if not rakip_row.empty else "Unknown"

            # durum
            durum = reverse_status

            # satırı ekle
            empty_idx = team_df[team_df['Rakip'] == ''].index.min()
            if pd.notna(empty_idx):
                team_df.loc[empty_idx] = {'Takım': team_name,'Ülke' : team_country,'Pot':pot_val,
                                          'Rakip': rakip, 'Rakip Ülke': rakip_ulke, 'Rakip Pot': 'Pot 2',
                                          'Durum': durum}
            else:
                team_df.loc[len(team_df)] = {'Takım': team_name,'Ülke' : team_country,'Pot':pot_val,
                                            'Rakip': rakip, 'Rakip Ülke': rakip_ulke, 'Rakip Pot': 'Pot 2',
                                            'Durum': durum}
    n = 2 - len(matches) if not matches.empty else 2

    # 2️⃣ Yeni eşleşmeler seç
    while n > 0 and not pot2_ave.empty:
        # 1️⃣ Önce pot_matches içinde bu takım var mı kontrol et
        # 1️⃣ Önce pot_matches içinde bu takım var mı kontrol et
        team_fix = pot_matches[(pot_matches['Takım'] == team_name) & (pot_matches['Rakip Pot'] == 'Pot 2')]
        opponent_fix = pot_matches[(pot_matches['Rakip'] == team_name) & (pot_matches['Pot'] == 'Pot 2')]
        opponent_pot=team_df[team_df['Rakip Pot'] == 'Pot 2']
        # team_df ve pot2_ave her satırda artık 'Ülke' sütununa sahip
        team_country = team_df['Ülke'].iloc[0]  # team_name'in ülkesi

        # Önce mevcut rakipleri ve rakip ülkelerini al
        existing_opponents = team_df['Rakip'].tolist()
        # Rakip olarak seçilmiş ülkelerin sayısını hesapla
        country_counts = team_df['Rakip Ülke'].value_counts()

        # 2'den fazla rakip olan ülkeleri blokla
        blocked_countries = [country for country, count in country_counts.items() if count >= 2]



        if not team_fix.empty:
            current_status = team_fix.iloc[0]['Durum']
            desired_status = 'Deplasman' if current_status == 'İç saha' else 'İç saha'

            valid_choices = pot2_ave[
                (pot2_ave['Takım'] != team_name) &
                (pot2_ave['Durum'] == desired_status) &
                (~pot2_ave['Takım'].isin(existing_opponents)) &
                (pot2_ave['Ülke'] != team_country) &
                (~pot2_ave['Ülke'].isin(blocked_countries))
            ]
        elif not opponent_fix.empty:
            current_status = opponent_fix.iloc[0]['Durum']
            desired_status = current_status

            valid_choices = pot2_ave[
                (pot2_ave['Takım'] != team_name) &
                (pot2_ave['Durum'] == desired_status) &
                (~pot2_ave['Takım'].isin(existing_opponents)) &
                (pot2_ave['Ülke'] != team_country) &
                (~pot2_ave['Ülke'].isin(blocked_countries))
            ]
        else:
            valid_choices = pot2_ave[
                (pot2_ave['Takım'] != team_name) &
                (~((pot2_ave['Durum'] == 'İç saha') & (opponent_pot['Durum'].eq('İç saha')).any())) &
                (~((pot2_ave['Durum'] == 'Deplasman') & (opponent_pot['Durum'].eq('Deplasman')).any())) &
                (~pot2_ave['Takım'].isin(existing_opponents)) &
                (pot2_ave['Ülke'] != team_country) &
                (~pot2_ave['Ülke'].isin(blocked_countries))
            ]
        if valid_choices.empty:
          print(f"⚠️ {team_name} için uygun rakip kalmadı.")
          raise RestartDraw  # Burada özel exception fırlatıyoruz

        idx = valid_choices.sample(n=1).index[0]
        selected = pot2_ave.loc[idx]

        rakip = selected['Takım']
        durum = selected['Durum']
        rakip_ulke = selected['Ülke']

        # İlk boş satıra yaz
        empty_idx = team_df[team_df['Rakip'] == ''].index.min()
        if pd.notna(empty_idx):
            team_df.loc[empty_idx] = {'Takım': team_name,'Ülke' : team_country,'Pot':pot_val,
                                      'Rakip': rakip, 'Rakip Ülke': rakip_ulke, 'Rakip Pot': 'Pot 2',
                                      'Durum': durum}
        else:
            team_df.loc[len(team_df)] = {'Takım': team_name,'Ülke' : team_country,'Pot':pot_val,
                                         'Rakip': rakip, 'Rakip Ülke': rakip_ulke, 'Rakip Pot': 'Pot 2',
                                         'Durum': durum}

        # pot_matches’e ekle
        pot_matches.loc[len(pot_matches)] = {'Takım': team_name,'Ülke' : team_country,'Pot':pot_val,
                                       'Rakip': rakip, 'Rakip Ülke': rakip_ulke, 'Rakip Pot': 'Pot 2',
                                       'Durum': durum}

        # Seçilen satırı sil
        pot2_ave.drop(idx, inplace=True)
        reverse_status = 'Deplasman' if durum == 'İç saha' else 'İç saha'
        pot2_ave = pot2_ave.drop(
            pot2_ave[(pot2_ave['Takım'] == team_name) & (pot2_ave['Durum'] == reverse_status)].index
        )

        n -= 1

def pot3_draw(team_name, team_df):
    global pot_matches, pot3_ave
    n = 2
    pot_val = team_df['Pot'].iloc[0]
    # Daha önce eklenmiş maçları bul
    matches = pot_matches[pot_matches['Rakip'] == team_name] if not pot_matches.empty else pd.DataFrame()
    if not matches.empty :
        matches = matches[matches['Pot'] == 'Pot 3']
        for idx, row in matches.iterrows():
            reverse_status = 'Deplasman' if row['Durum'] == 'İç saha' else 'İç saha'

            # team_country'i güvenli al
            team_row = team_df.loc[team_df['Takım'] == team_name, 'Ülke']
            team_country = team_row.iloc[0] if not team_row.empty else "Unknown"

            # rakip ve rakip_ulke (pot_matches'ten alınıyor)
            rakip = row['Takım']
            rakip_row = pot_matches.loc[pot_matches['Takım'] == rakip, 'Ülke']
            rakip_ulke = rakip_row.iloc[0] if not rakip_row.empty else "Unknown"


            # durum
            durum = reverse_status

            # satırı ekle
            empty_idx = team_df[team_df['Rakip'] == ''].index.min()
            if pd.notna(empty_idx):
                team_df.loc[empty_idx] = {'Takım': team_name,'Ülke' : team_country,'Pot':pot_val,
                                          'Rakip': rakip, 'Rakip Ülke': rakip_ulke, 'Rakip Pot': 'Pot 3',
                                          'Durum': durum}
            else:
                team_df.loc[len(team_df)] = {'Takım': team_name,'Ülke' : team_country,'Pot':pot_val,
                                            'Rakip': rakip, 'Rakip Ülke': rakip_ulke, 'Rakip Pot': 'Pot 3',
                                            'Durum': durum}
    n = 2 - len(matches) if not matches.empty else 2

    # 2️⃣ Yeni eşleşmeler seç
    while n > 0 and not pot3_ave.empty:
        # 1️⃣ Önce pot_matches içinde bu takım var mı kontrol et
        team_fix = pot_matches[(pot_matches['Takım'] == team_name) & (pot_matches['Rakip Pot'] == 'Pot 3')]
        opponent_fix = pot_matches[(pot_matches['Rakip'] == team_name) & (pot_matches['Pot'] == 'Pot 3')]
        opponent_pot=team_df[team_df['Rakip Pot'] == 'Pot 3']
        # team_df ve pot3_ave her satırda artık 'Ülke' sütununa sahip
        team_country = team_df['Ülke'].iloc[0]  # team_name'in ülkesi

        # Önce mevcut rakipleri ve rakip ülkelerini al
        existing_opponents = team_df['Rakip'].tolist()
        # Rakip olarak seçilmiş ülkelerin sayısını hesapla
        country_counts = team_df['Rakip Ülke'].value_counts()

        # 2'den fazla rakip olan ülkeleri blokla
        blocked_countries = [country for country, count in country_counts.items() if count >= 2]

        if not team_fix.empty:
            current_status = team_fix.iloc[0]['Durum']
            desired_status = 'Deplasman' if current_status == 'İç saha' else 'İç saha'

            valid_choices = pot3_ave[
                (pot3_ave['Takım'] != team_name) &
                (pot3_ave['Durum'] == desired_status) &
                (~pot3_ave['Takım'].isin(existing_opponents)) &
                (pot3_ave['Ülke'] != team_country) &
                (~pot3_ave['Ülke'].isin(blocked_countries))
            ]

        elif not opponent_fix.empty:
            current_status = opponent_fix.iloc[0]['Durum']
            desired_status = current_status

            valid_choices = pot3_ave[
                (pot3_ave['Takım'] != team_name) &
                (pot3_ave['Durum'] == desired_status) &
                (~pot3_ave['Takım'].isin(existing_opponents)) &
                (pot3_ave['Ülke'] != team_country) &
                (~pot3_ave['Ülke'].isin(blocked_countries))
            ]

        else:
            valid_choices = pot3_ave[
                (pot3_ave['Takım'] != team_name) &
                (~pot3_ave['Takım'].isin(existing_opponents)) &
                (~((pot3_ave['Durum'] == 'İç saha') & (opponent_pot['Durum'].eq('İç saha')).any())) &
                (~((pot3_ave['Durum'] == 'Deplasman') & (opponent_pot['Durum'].eq('Deplasman')).any())) &
                (pot3_ave['Ülke'] != team_country) &
                (~pot3_ave['Ülke'].isin(blocked_countries))
            ]

        if valid_choices.empty:
          print(f"⚠️ {team_name} için uygun rakip kalmadı.")
          raise RestartDraw  # Burada özel exception fırlatıyoruz

        idx = valid_choices.sample(n=1).index[0]
        selected = pot3_ave.loc[idx]

        rakip = selected['Takım']
        durum = selected['Durum']
        rakip_ulke = selected['Ülke']

        # İlk boş satıra yaz
        empty_idx = team_df[team_df['Rakip'] == ''].index.min()
        if pd.notna(empty_idx):
            team_df.loc[empty_idx] = {'Takım': team_name,'Ülke' : team_country,'Pot':pot_val,
                                      'Rakip': rakip, 'Rakip Ülke': rakip_ulke, 'Rakip Pot': 'Pot 3',
                                      'Durum': durum}
        else:
            team_df.loc[len(team_df)] = {'Takım': team_name,'Ülke' : team_country,'Pot':pot_val,
                                         'Rakip': rakip, 'Rakip Ülke': rakip_ulke, 'Rakip Pot': 'Pot 3',
                                         'Durum': durum}

        # pot_matches’e ekle
        pot_matches.loc[len(pot_matches)] = {'Takım': team_name,'Ülke' : team_country,'Pot':pot_val,
                                       'Rakip': rakip, 'Rakip Ülke': rakip_ulke, 'Rakip Pot': 'Pot 3',
                                       'Durum': durum}

        # Seçilen satırı sil
        pot3_ave.drop(idx, inplace=True)
        reverse_status = 'Deplasman' if durum == 'İç saha' else 'İç saha'
        pot3_ave = pot3_ave.drop(
            pot3_ave[(pot3_ave['Takım'] == team_name) & (pot3_ave['Durum'] == reverse_status)].index
        )

        n -= 1

def pot4_draw(team_name, team_df):
    global pot_matches, pot4_ave
    n = 2
    pot_val = team_df['Pot'].iloc[0]
    # Daha önce eklenmiş maçları bul
    matches = pot_matches[pot_matches['Rakip'] == team_name] if not pot_matches.empty else pd.DataFrame()
    if not matches.empty :
        matches = matches[matches['Pot'] == 'Pot 4']
        for idx, row in matches.iterrows():
            reverse_status = 'Deplasman' if row['Durum'] == 'İç saha' else 'İç saha'

            # team_country'i güvenli al
            team_row = team_df.loc[team_df['Takım'] == team_name, 'Ülke']
            team_country = team_row.iloc[0] if not team_row.empty else "Unknown"
            opponent_pot=team_df[team_df['Rakip Pot'] == 'Pot 4']
            # rakip ve rakip_ulke (pot_matches'ten alınıyor)
            rakip = row['Takım']
            rakip_row = pot_matches.loc[pot_matches['Takım'] == rakip, 'Ülke']
            rakip_ulke = rakip_row.iloc[0] if not rakip_row.empty else "Unknown"

            # durum
            durum = reverse_status

            # satırı ekle
            empty_idx = team_df[team_df['Rakip'] == ''].index.min()
            if pd.notna(empty_idx):
                team_df.loc[empty_idx] = {'Takım': team_name,'Ülke' : team_country,'Pot':pot_val,
                                          'Rakip': rakip, 'Rakip Ülke': rakip_ulke, 'Rakip Pot': 'Pot 4',
                                          'Durum': durum}
            else:
                team_df.loc[len(team_df)] = {'Takım': team_name,'Ülke' : team_country,'Pot':pot_val,
                                            'Rakip': rakip, 'Rakip Ülke': rakip_ulke,'Rakip Pot': 'Pot 4',
                                            'Durum': durum}
    n = 2 - len(matches) if not matches.empty else 2

    # 2️⃣ Yeni eşleşmeler seç
    while n > 0 and not pot4_ave.empty:
        # Önce pot_matches içinde bu takım var mı kontrol et
        team_fix = pot_matches[(pot_matches['Takım'] == team_name) & (pot_matches['Rakip Pot'] == 'Pot 4')]
        opponent_fix = pot_matches[(pot_matches['Rakip'] == team_name) & (pot_matches['Pot'] == 'Pot 4')]
        opponent_pot=team_df[team_df['Rakip Pot'] == 'Pot 4']
        team_country = team_df['Ülke'].iloc[0]  # team_name'in ülkesi

        # Mevcut rakipler ve rakip ülkeleri
        existing_opponents = team_df['Rakip'].tolist()
        # Rakip olarak seçilmiş ülkelerin sayısını hesapla
        country_counts = team_df['Rakip Ülke'].value_counts()

        # 2'den fazla rakip olan ülkeleri blokla
        blocked_countries = [country for country, count in country_counts.items() if count >= 2]

        if not team_fix.empty:
            current_status = team_fix.iloc[0]['Durum']
            desired_status = 'Deplasman' if current_status == 'İç saha' else 'İç saha'

            valid_choices = pot4_ave[
                (pot4_ave['Takım'] != team_name) &
                (pot4_ave['Durum'] == desired_status) &
                (~pot4_ave['Takım'].isin(existing_opponents)) &
                (pot4_ave['Ülke'] != team_country) &
                (~pot4_ave['Ülke'].isin(blocked_countries))
            ]

        elif not opponent_fix.empty:
            current_status = opponent_fix.iloc[0]['Durum']
            desired_status = current_status

            valid_choices = pot4_ave[
                (pot4_ave['Takım'] != team_name) &
                (pot4_ave['Durum'] == desired_status) &
                (~pot4_ave['Takım'].isin(existing_opponents)) &
                (pot4_ave['Ülke'] != team_country) &
                (~pot4_ave['Ülke'].isin(blocked_countries))
            ]

        else:
            valid_choices = pot4_ave[
                (pot4_ave['Takım'] != team_name) &
                (~pot4_ave['Takım'].isin(existing_opponents)) &
                (~((pot4_ave['Durum'] == 'İç saha') & (opponent_pot['Durum'].eq('İç saha')).any())) &
                (~((pot4_ave['Durum'] == 'Deplasman') & (opponent_pot['Durum'].eq('Deplasman')).any())) &
                (pot4_ave['Ülke'] != team_country) &
                (~pot4_ave['Ülke'].isin(blocked_countries))
            ]

        if valid_choices.empty:
          print(f"⚠️ {team_name} için uygun rakip kalmadı.")
          raise RestartDraw  # Burada özel exception fırlatıyoruz

        idx = valid_choices.sample(n=1).index[0]
        selected = pot4_ave.loc[idx]

        rakip = selected['Takım']
        durum = selected['Durum']
        rakip_ulke = selected['Ülke']

        # İlk boş satıra yaz
        empty_idx = team_df[team_df['Rakip'] == ''].index.min()
        if pd.notna(empty_idx):
            team_df.loc[empty_idx] = {'Takım': team_name,'Ülke' : team_country,'Pot':pot_val,
                                      'Rakip': rakip, 'Rakip Ülke': rakip_ulke,'Rakip Pot': 'Pot 4',
                                      'Durum': durum}
        else:
            team_df.loc[len(team_df)] = {'Takım': team_name,'Ülke' : team_country,'Pot':pot_val,
                                         'Rakip': rakip, 'Rakip Ülke': rakip_ulke,'Rakip Pot': 'Pot 4',
                                         'Durum': durum}

        # pot_matches’e ekle
        pot_matches.loc[len(pot_matches)] = {'Takım': team_name,'Ülke' : team_country,'Pot':pot_val,
                                       'Rakip': rakip, 'Rakip Ülke': rakip_ulke,'Rakip Pot': 'Pot 4',
                                       'Durum': durum}

        # Seçilen satırı sil
        pot4_ave.drop(idx, inplace=True)
        reverse_status = 'Deplasman' if durum == 'İç saha' else 'İç saha'
        pot4_ave = pot4_ave.drop(
            pot4_ave[(pot4_ave['Takım'] == team_name) & (pot4_ave['Durum'] == reverse_status)].index
        )

        n -= 1

def draw(df):
    created_dfs = []  # Üretilen tüm dataframe isimlerini saklamak için

    # ÖNEMLİ: Pot DataFrame'lerini sadece BİR KEZ oluştur
    create_pot_ave_from_df(df, ulkeler)

    # Pot 1'deki takımlar
    for team, country in zip(df['Pot 1'].iloc[0:9].tolist(), ulkeler[0:9]):
        var_name = f"df_{team.replace(' ', '_').replace('-', '_')}"
        globals()[var_name] = pd.DataFrame({
            'Takım': [team]*8,
            'Ülke': [country]*8,
            'Pot' : ['Pot 1']*8,
            'Rakip': ['']*8,
            'Rakip Ülke': ['']*8,
            'Rakip Pot' : ['']*8,
            'Durum': ['']*8
        })
        print(f"Created dataframe: {var_name}")
        created_dfs.append(var_name)

        # Pot draw fonksiyonları
        pot1_draw(team, globals()[var_name])
        pot2_draw(team, globals()[var_name])
        pot3_draw(team, globals()[var_name])
        pot4_draw(team, globals()[var_name])

    create_pot_ave_from_df(df, ulkeler)  # BU SATIRI KALDIR!

    # Pot 2'deki takımlar
    for team, country in zip(df['Pot 2'].iloc[0:9].tolist(), ulkeler[9:18]):
        var_name = f"df_{team.replace(' ', '_').replace('-', '_')}"
        globals()[var_name] = pd.DataFrame({
            'Takım': [team]*8,
            'Ülke': [country]*8,
            'Pot' : ['Pot 2']*8,
            'Rakip': ['']*8,
            'Rakip Ülke': ['']*8,
            'Rakip Pot' : ['']*8,
            'Durum': ['']*8
        })
        print(f"Created dataframe: {var_name}")
        created_dfs.append(var_name)

        # Pot draw fonksiyonları
        pot1_draw(team, globals()[var_name])
        pot2_draw(team, globals()[var_name])
        pot3_draw(team, globals()[var_name])
        pot4_draw(team, globals()[var_name])

    create_pot_ave_from_df(df, ulkeler)  # BU SATIRI KALDIR!

    # Pot 3'deki takımlar
    for team, country in zip(df['Pot 3'].iloc[0:9].tolist(), ulkeler[18:27]):
        var_name = f"df_{team.replace(' ', '_').replace('-', '_')}"
        globals()[var_name] = pd.DataFrame({
            'Takım': [team]*8,
            'Ülke': [country]*8,
            'Pot' : ['Pot 3']*8,
            'Rakip': ['']*8,
            'Rakip Ülke': ['']*8,
            'Rakip Pot' : ['']*8,
            'Durum': ['']*8,
        })
        print(f"Created dataframe: {var_name}")
        created_dfs.append(var_name)

        # Pot draw fonksiyonları
        pot1_draw(team, globals()[var_name])
        pot2_draw(team, globals()[var_name])
        pot3_draw(team, globals()[var_name])
        pot4_draw(team, globals()[var_name])

    create_pot_ave_from_df(df, ulkeler)  # BU SATIRI KALDIR!

    # Pot 4'teki takımlar
    for team, country in zip(df['Pot 4'].iloc[0:9].tolist(), ulkeler[27:36]):
        var_name = f"df_{team.replace(' ', '_').replace('-', '_')}"
        globals()[var_name] = pd.DataFrame({
            'Takım': [team]*8,
            'Ülke': [country]*8,
            'Pot' : ['Pot 4']*8,
            'Rakip': ['']*8,
            'Rakip Ülke': ['']*8,
            'Rakip Pot' : ['']*8,
            'Durum': ['']*8
        })
        print(f"Created dataframe: {var_name}")
        created_dfs.append(var_name)

        # Pot draw fonksiyonları
        pot1_draw(team, globals()[var_name])
        pot2_draw(team, globals()[var_name])
        pot3_draw(team, globals()[var_name])
        pot4_draw(team, globals()[var_name])


    # Tüm dataframe'leri birleştir
    df_fixtures = pd.concat([globals()[name] for name in created_dfs], ignore_index=True)
    return df_fixtures
```
